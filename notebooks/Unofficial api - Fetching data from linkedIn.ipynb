```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "carbide-haven-450821-t3-2b90ba098b11.json"


from google.cloud import firestore

# Initialize Firestore client
db = firestore.Client()

print("✅ Connected to Firestore!")
```

```python
from linkedin_api import Linkedin

# Replace with your LinkedIn email and password, create a second accound, as LinkedIn may block your account
api = Linkedin("kuxki631@gmail.com", "Link20250203")
```

```python

import time  # Optional: To add delay if needed


# Reference Firestore collection
raw_jobs_ref = db.collection("raw_job_data")


print("🔄 Fetching job listings...")

# Call the API to get job listings
job_listings = api.search_jobs('Software Engineer')
```

```python
if not job_listings:
    print("⚠️ No job listings found.")
```

```python
batch = db.batch()
batch_size = 0
new_jobs_count = 0

for job in job_listings:
    # Extract job ID from 'entityUrn'
    job_id = job["entityUrn"].split(":")[-1]
    job_doc_id = f"job_{job_id}"  # Use job_<id> as Firestore document ID

    # Check if the job already exists
    if raw_jobs_ref.document(job_doc_id).get().exists:
        print(f"✅ Job {job_id} already exists. Skipping.")
        continue

    # Add job to Firestore batch
    job_ref = raw_jobs_ref.document(job_doc_id)
    batch.set(job_ref, job)

    batch_size += 1
    new_jobs_count += 1

    # Commit batch every 500 writes (Firestore batch limit)
    if batch_size >= 500:
        batch.commit()
        batch = db.batch()  # Start a new batch
        batch_size = 0  # Reset batch size
        print("🚀 Committed a batch of 500 jobs to Firestore.")

# Commit any remaining jobs
if batch_size > 0:
    batch.commit()
    print(f"🚀 Committed final batch of {batch_size} jobs to Firestore.")

print(f"🎉 Job search results updated. {new_jobs_count} new jobs saved.")
```
